**뷰티풀 수프로 일별 시세 읽어오기**

```python
# 페이지 수 추출
# 책 내용대로 urlib 적용 불가, requests 적용

import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = 'https://finance.naver.com/item/sise_day.naver?code=034020&page=1'
headers = {'User-Agent': 'Mozilla/5.0'}
r = requests.get(url, headers = headers)
soup = BeautifulSoup(r.text, 'html.parser')
tag = soup.select_one('td.pgRR > a')['href']
sp = tag.split('=')
print(sp)
last_page = sp[-1]
print('맨 뒤 페이지 번호 : ', last_page)
```

```python
# 전체 페이지 읽어오기
df = pd.DataFrame()
sise_url = 'https://finance.naver.com/item/sise_day.naver?code=034020'

for page in range(1, int(last_page)+1):
    page_url = '{}&page={}'.format(sise_url, page)
    df = df.append(pd.read_html(requests.get(page_url, headers={'User-agent': 'Mozilla/5.0'}).text)[0])

df = df.dropna()
print(df)
```

### 02.2 OHLC와 캔들 차트

**두산중공업 종가 차트**

```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
from matplotlib import pyplot as plt

df = df.iloc[0:30] # 데이터가 너무 많으므로, 최근 30일 활용
df = df.sort_values(by='날짜') # 날짜 오름차순 정렬

# 날짜, 종가 칼럼으로 차트 그리기
plt.figure(figsize = (15,8))
plt.title('Doosan Heavy Industries & Construction (close)')
plt.xticks(rotation=45)
plt.plot(df['날짜'], df['종가'], 'co-')
plt.grid(color='gray', linestyle='--')
plt.show()
```

**두산중공업 캔들 차트**
- 구버전(mpl_finance) 패키지에서 신버전(mplfinance) 패키지로 새롭게 변경되었다.
- 설치 : pip install mpl_finance > pip install --upgrade mplfinance
> import mplfinance as mpf 
>
> mpf.plot(OHLC 데이터프레임 [, title=차트제목][, type=차트형태]
>
>          [, mav=이동평균선][, volume=거래량 표시여부][, ylabel=y축 레이블])

```python
### 신버전, 데이터 추출-가공 코드 새로 작성
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import mplfinance as mpf

# 맨 뒤 페이지 수 구하기
url = 'https://finance.naver.com/item/sise_day.naver?code=034020&page=1'
headers = {'User-Agent': 'Mozilla/5.0'}
r = requests.get(url, headers = headers)
soup = BeautifulSoup(r.text, 'html.parser')
tag = soup.select_one('td.pgRR > a')['href']
sp = tag.split('=')
last_page = sp[-1]

# 전체 페이지 추출
df = pd.DataFrame()
sise_url = 'https://finance.naver.com/item/sise_day.naver?code=034020'

for page in range(1, int(last_page)+1):
    page_url = '{}&page={}'.format(sise_url, page)
    df = df.append(pd.read_html(requests.get(page_url, headers={'User-agent': 'Mozilla/5.0'}).text)[0])

# 차트 출력을 위한 데이터프레임 가공
df = df.dropna()
df = df.iloc[0:30]
df = df.rename(columns={'날짜':'Date', '시가':'Open', '고가':'High', '저가':'Low', '종가':'Close', '거래량':'Volume'})
df = df.sort_values(by='Date')
df.index = pd.to_datetime(df.Date)
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
pd.read_html(requests.get(page_url, headers={'User-agent': 'Mozilla/5.0'}).text)[0]
```

```python
# mplfinance로 캔들 차트 그리기
mpf.plot(df, title='Doosan Heavy Industries & Construction : Candle Chart', type='candle')
```

```python
# OHLC 차트 그리기
mpf.plot(df, title='Doosan Heavy Industries & Construction : OHLC Chart')
```

```python
# 캔들 차트 디자인 변경
# kwargs : keyword arguments
kwargs = dict(title = 'Doosan Heavy Industries & Construction : Customized Candle Chart',
             type = 'candle', mav=(2,4,6), volume=True, ylabel='ohlc candles')
mc = mpf.make_marketcolors(up='royalblue', down='tomato', inherit=True)
s = mpf.make_mpf_style(marketcolors=mc)
mpf.plot(df, **kwargs, style=s)
```
