### 01.2 일간 변동률로 주가 비교하기

> $R_t(오늘 변동률) = \Big({\cfrac{R_t(오늘 종가)-R_{t-1}(어제 종가)\quad\quad}{\qquad R_{t-1}(어제 종가)\qquad\quad}}\Big)\times 100$

- 시리즈 모듈에서 제공하는 shift() 함수를 사용해 위의 식 적용

```python
# 삼성전자 종가 칼럼의 데이터 확인
print(type(sec['Close'])) # 객체 타입 확인 : series
print()
print(sec['Close'])
```

```python
# '어제'의 종가 구하기
sec['Close'].shift(1)
```

```python
# 일간 변동률 구하기
sec_dpc = (sec['Close']/sec['Close'].shift(1) - 1)*100
sec_dpc.iloc[0] = 0 #첫 번째 일간 변동률 값 NaN을 0으로 변환
sec_dpc.head()
```

- bins : 구간 수 (hist()함수에서 사용되는 default = 10)

> 해석 : 0 bin을 기준으로 좌우 대칭적인 그래프.

>
> 급첨 분포(leptokurtic distribution) : 정규분포와 비교했을 때, 주가의 움직임이 대부분 매우 작은 범위 안에서 발생함을 의미
>
> 팻 테일(fat tail) : 그래프의 좌우 극단 부분에 해당하는 아주 큰 가격 변동이 정규분포보다 더 많이 발생함을 의미

## 01.4 일간 변동률 누적합 구하기
- Series의 cumsum() 함수 활용

```python
sec_dpc_cs = sec_dpc.cumsum()
sec_dpc_cs
```

> 조회기간 동안 누적 77% 수익률 기록

**삼성전자와 마이크로소프트 주식수익률 비교 시각화**

```python
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()

# 삼성전자
sec = pdr.get_data_yahoo('005930.KS', start='2019-01-01')
sec_dpc = (sec['Close']-sec['Close'].shift(1)) / sec['Close'].shift(1) * 100
sec_dpc.iloc[0] = 0 # 일간 변동률 첫 번째 값인 NaN 0으로 변환
sec_dpc_cs = sec_dpc.cumsum() # 일간 변동률의 누적합

# Microsoft
msft = pdr.get_data_yahoo('MSFT', start='2019-01-01')
msft_dpc = (msft['Close']-msft['Close'].shift(1)) / msft['Close'].shift(1) * 100
msft_dpc.iloc[0] = 0
msft_dpc_cs = msft_dpc.cumsum()

import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.plot(sec_dpc_cs.index, sec_dpc_cs, 'b', label='Samsung Electronics')
plt.plot(msft_dpc_cs.index, msft_dpc_cs, 'r--', label='Microsoft')
plt.ylabel('Change %')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

## 02 최대 손실 낙폭
> 최대손실낙폭(MDD, Maximum Drawdown) : 특정 기간 발생한 최고점에서 최저점까지의 가장 큰 손실
> 
> 수익률을 높이는 것보다 MDD를 낮추는 것이 나은 경우가 많을 만큼 중요한 지표
> 
> $MDD = \cfrac{최저점 - 최고점}{최저점}$

### 02.1 KOSPI의 MDD
- rolling() : 시리즈에서 윈도우 크기에 해당하는 개수만큼 데이터를 추출하여 집계 함수에 해당하는 연산 실시
- min_periods를 지정하면 데이터 개수가 윈도우 크기에 못미치더라도 지정한 개수만 만족하면 연산 수행

```python
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
import matplotlib.pyplot as plt

kospi = pdr.get_data_yahoo('^KS11', '2004-01-04') # KOSPI 지수의 심볼 : ^KS11

window = 252 # window 값은 1년 동안의 개장일을 252일로 어림해 설정
peak = kospi['Adj Close'].rolling(window, min_periods=1).max() # KOSPI 종가 칼럼에서 1년(거래일 기준) 기간 단위로 최고치 peak 구하기
drawdown = kospi['Adj Close']/peak - 1.0 # drawdown은 최고치(peak) 대비 현재 KOSPI 종가가 얼마나 하락했는지 구한다.
max_dd = drawdown.rolling(window, min_periods = 1).min()
# drawdown에서 1년 기간 단위로 최저치 max_dd를 구한다. 마이너스값이므로, 최저치가 최대 손실 낙폭

plt.figure(figsize=(9, 7))
plt.subplot(211)
kospi['Close'].plot(label='KOSPI', title='KOSPI MDD', grid=True, legend=True)
plt.subplot(212)
drawdown.plot(c='blue', label='KOSPI DD', grid=True, legend=True)
max_dd.plot(c='red', label='KOSPI MDD', grid=True, legend=True)
plt.show()
```

```python
print('전체 기간 중 최대손실낙폭이 최대인 값 찾기 : 서브프라임 모기지 영향')
print(max_dd.min())
print()
print('2020년 이후 최대손실낙폭이 최대인 값 찾기 : 코로나 영향')
print(max_dd[max_dd.index >= '2020-01-01'].min())
```

```python
print('서브프라임 모기지 사태 직후 최대 손실 낙폭 발생일 : ')
print(max_dd[max_dd==-0.5453665130144085])
print('코로나 이후 최대 손실 낙폭 발생일 : ')
print(max_dd[max_dd==-0.3570889779916474])
```

### 03.5 선형 회귀 분석
- 회귀 모델 : 연속적인 데이터 Y와 Y의 원인이 되는 X 간의 관계를 추정하는 관계식
- 측정 상 한계로 인한 잡음(noise)이 존재하기 때문에, 정확한 관계식을 표현하는 확률 변수인 오차항을 둔다.
- 독립변수 X와 종속변수 Y의 관계가 1차식으로 나타날 때 선형 회귀 모델(linear regression model)이라고 부른다.
> $Y_i = \beta_0 + \beta_1X_i + \epsilon_i(i=1,2,...,n)$
>
> - $Y_i$ : i번째 종속변수의 값
> - $X_i$ : i번째 독립변수의 값
> - $\beta_0$ : 선형 회귀식의 절편(intercept)
> - $\beta_1$ : 선형 회귀식의 기울기(slope)
> - $\epsilon_i$ : 오차항(종속변수 Y의 실제값과 기대치의 차이)
>
> Y의 기대치 $E(Y) = \beta_0 + \beta_1X_1$ -> 회귀 함수(regression fuction)

- stats : Scipy 패키지의 서브 패키지. 다양한 통계 함수 제공 -> linregress() 사용

```python
from scipy import stats
regr = stats.linregress(df['DOW'], df['KOSPI'])
regr
```

> 해석 : slope(기울기), intercept(y절편), rvalue(r값, 상관계수), pvalue(p값), stderr(표준편차)

## 04. 상관계수에 따른 리스크 완화
- 상관계수(Coefficient of Correlation) : 독립-종속변수 사이 상관관계의 정도를 나타내는 수치
- 항상 -1 <= r <= 1을 만족하며, 양의 상관관계가 강한 값을 1, 음의 상관관계가 강한 값을 -1로 표현

### 04.1 데이터프레임으로 상관계수 구하기

```python
df.corr()
```

### 04.2 시리즈로 상관계수 구하기

```python
df['DOW'].corr(df['KOSPI'])
```

### 04.3 결정계수 구하기
- 결정계수(R-squared) : 회귀선이 실제로 데이터를 어느 정도 설명하는지를 나타내는 계수. 상관계수(R value)의 제곱


```python
r_value = df['DOW'].corr(df['KOSPI'])
r_squared = r_value ** 2
print('상관계수 : {}, 결정계수 : {}'.format(r_value, r_squared))
```

> 해석 : 결정계수가 1이면 모든 표본 관측치가 추정된 회귀선 상에만 있다는 의미로, 변수 간 관계가 완벽히 설명된다. 반면 결정계수가 0이면 추정된 회귀선이 변수 사이 관계를 전혀 설명하지 못한다는 의미다.

### 04.4 다우존스 지수와 KOSPI의 회귀 분석
- 회귀 분석 후 회귀선을 시각화해 분석하기
- 전체 소스 코드를 다시 작성했다.

```python
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()

dow = pdr.get_data_yahoo('^DJI', '2000-01-14')
kospi = pdr.get_data_yahoo('^KS11', '2000-01-14')

df = pd.DataFrame({'X' : dow['Close'], 'Y' : kospi['Close']})
df = df.fillna(method='bfill')
df = df.fillna(method='ffill')

regr = stats.linregress(df.X, df.Y)
regr_line = f'Y = {regr.slope:.2f} * X + {regr.intercept:.2f}'

plt.figure(figsize=(7, 7))
plt.plot(df.X, df.Y, '.')
plt.plot(df.X, regr.slope * df.X + regr.intercept, 'r')
plt.legend(['DOW x KOSPI', regr_line])
plt.title(f'DOW x KOSPI (R = {regr.rvalue:.2f})')
plt.xlabel('Dow Jones Industrial Average')
plt.ylabel('KOSPI')
plt.show()
```

### 04.6 상관계수에 따른 리스크 완화
> Harry Markowitz의 modern portfolio theory(MPT) : 
>
>'투자에 대한 수익과 위험은 평균과 분산으로 나타낼 수 있으며, 상관관계가 낮은 자산을 대상으로 분산 투자하면 위험을 감소시킬 수 있다.
> 
> 상관계수에 따른 리스크 완화 효과
> - +1.0 : 리스크 완화 효과 없음
> - +0.5 : 중간 정도의 리스크 완화 효과가 있음
> - 0 : 상당한 리스크 완화 효과가 있음
> - -0.5 : 대부분의 리스크를 제거함
> - -1.0 : 모든 리스크 제거함
